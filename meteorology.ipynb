```python
import requests
import json
import pandas as pd
from datetime import datetime

meteo = requests.get('https://www.meteoschweiz.admin.ch/product/input/measured-values/messwerte-lufttemperatur-10min/stationsTable/stationsTable.messwerte-lufttemperatur-10min.de.json').json()

data={
        
        'Station': [],
         'Abk.' :[],
         'Temperatur': [],
         'Messdatum' : [],
         'Messhöhe' : [],
         'Zusatzinformation': [],
         'CH Koordinaten' : [],
         'Breiten-/Längengrad' : [],
         'Kanton' : []
         
         }

for i in meteo['stations']:
    data['Station'].append(i['station_name'])
    data['Abk.'].append(i['id'])
    data['Temperatur'].append(i['current']['value'])
    data['Messdatum'].append(datetime.fromtimestamp(i['current']['date']/1000))
    data['Messhöhe'].append(i['altitude'])
    data['Zusatzinformation'].append(i['measurement_height'])
    data['CH Koordinaten'].append((str(i['coordinates'][0])+" - "+str(i['coordinates'][1])))
    data['Breiten-/Längengrad'].append(str(i['latlong'][0])+" - "+str(i['latlong'][1]))
    data['Kanton'].append(i['canton'])
    
      

df = pd.DataFrame(data)
print(df[['Station', 'Abk.','Temperatur','Messdatum','Messhöhe','Zusatzinformation','CH Koordinaten','Breiten-/Längengrad','Kanton']]) 
```
