```python
import requests
from bs4 import BeautifulSoup
from urllib3.exceptions import ProtocolError
import csv


def get_page(url):
    response=requests.get(url)
    if not response.ok:
        print('Server responded:', response.status_code)
    else:
        soup=BeautifulSoup(response.text,'lxml')
    return soup

def get_detail_data(soup):
    #title
    #price
    #location
    #shipping
    #sellername
    #condition
    #color
    try:
        try:
            title = soup.find('h1', id='itemTitle').text.strip().replace('Details about  \xa0', '')
        except:
            title=soup.find('h3',attrs={'class':'s-item__title s-item__title--has-tags'}).text.strip().replace('Details about  \xa0', '')
    except:
        title=''
    print(title)
    
    try:
        try:
            p=soup.find('span', id='prcIsum').text.strip()
        except:
            p=soup.find('span', id='mm-saleDscPrc').text.strip()
        currency, price = p.split(' ')
    except:
        currency=''
        price=''
    print(price)
    print(currency)
    try:
        location = soup.find('div', attrs = {'class':'iti-eu-bld-gry'}).get_text().strip()
    except:
        location=''
    
    try:
        try:
            shipping_cost=soup.find('span', attrs={'id':'fshippingCost'}).get_text().strip()
        except:
            shipping_cost=soup.find('span', attrs={'id':'convetedPriceId'}).get_text().strip()
    except:
        shipping_cost= ''
    
    try:
        sellername = soup.find('span', attrs = {'class':'mbg-nw'}).get_text().strip()
    except:
        sellername = ''
    
    try:
        try:
            condition = soup.find('div', attrs = {'class':"u-flL condText  "}).get_text().strip()
        except:
            condition=soup.finf('div',attrs={'id':"vi-itm-cond"}).get_text().strip()
    except:
        condition= ''
    
    try:
        color=soup.find('select', attrs={'id':'msku-sel-1'}).get_text().strip().replace('- Select -', '')
    except:
        color= ''
    
    
    data={
        'title': title,
        'price': price,
        'location':location,
        'shipping_cost':shipping_cost,
        'sellername':sellername,
        'condition':condition,
        'color':color
        
    }
    
    return data


        
def get_index_data(soup):
    try:
        links=soup.find_all('a', class_='s-item__link')
    except:
        links=[]
        
    urls=[item.get('href') for item in links]
    
    return urls

import time

time.sleep(2.4)
print("Printed after 2.4 seconds.")

def write_csv(data, url):
    with open('output14032020.csv', 'a', encoding='utf-8') as csvfile:
        writer=csv.writer(csvfile)
        
        row=[data['title'], data['price'],data['location'],data['shipping_cost'],data['sellername'],data['condition'],data['color'],url]
    
        writer.writerow(row)

def main():
    
    url="https://www.ebay.com/sch/i.html?_nkw=robotic+vacuum+cleaner&_pgn={}"
    page_num = 1
    while page_num<=50:
        response = requests.get(url.format(page_num))
        if response.status_code == 404: # break once the page is not found
            break
        print("Scraping Page Number {}".format(page_num))
        soup = BeautifulSoup(response.text,"lxml")
    
        page_num += 1
  
    products= get_index_data(get_page(url))
    
    for link in products:
        data=get_detail_data(get_page(link))
        write_csv(data,link)
    


if __name__=='__main__':
    main()
```
