## WEB SCRAPİNG WİTH PYTHON USİNG BEAUTİFULSOUP FROM AMAZON


```python
import pandas as pd
import numpy as np
import requests
import time
from bs4 import BeautifulSoup as bs
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}
```

```python
for pagenumber in range(1,5):
    url = 'https://www.amazon.com/s?k=robot+vacuum+cleaner&page=' + str(pagenumber)
    r=requests.get(url, headers=headers)
    soup=bs(r.content, features="lxml")
    products=soup.find_all("div",attrs={"class":"s-include-content-margin s-border-bottom"})
    l_pn = []
    l_sb = []
    l_p  = []
    l_e  = []
    l_c  = []
    for product in products:
        product_link="https://www.amazon.com"+product.a.get("href")
        time.sleep(2.4)
        print("Printed after 2.4 seconds.")
        print(product_link)
        try:
            product_r=requests.get(product_link, headers=headers)
        except:
            print("none product details")
        product_soup=bs(product_r.content)
        features=product_soup.find_all("div", attrs={"class":"centerColAlign"})
        for feature in features:
            try:
                product_name=feature.select("#productTitle")[0].get_text().strip()
                l_pn.append(product_name)
            except:
                l_pn.append(np.nan)
            try:
                sold_by=feature.select("#bylineInfo")[0].get_text().strip()
                l_sb.append(sold_by)
            except:
                l_p.append(np.nan)
            try:
                price=feature.select("#priceblock_ourprice")[0].get_text().strip()
                l_p.append(price)
            except:
                l_p.append(np.nan)
            try:
                edition=feature.select(".selection")[1].get_text().strip()
                l_e.append(edition)
            except:
                l_e.append(np.nan)
            try:
                color=feature.select(".selection")[0].get_text().strip()
                l_c.append(color)
            except:
                l_c.append(np.nan)
            print("{}:{}:{}:{}:{}".format(product_name,sold_by,price,edition,color))
```
